```python
## IMPORT PACKAGES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import os, csv
from scipy import stats
import statsmodels.formula.api as smf
from mpl_toolkits.mplot3d import Axes3D


%matplotlib inline
```

```python
## LOAD DATA

# set working directory
hobo_data = r'/mnt/c/Users/MGLoso/Documents/software/KennGlacHobos'
# Load an excel workbook that is in the local directory
xlsx=pd.ExcelFile('KennGlacHobosData.xlsx')
# Load worksheet names
sheets=xlsx.sheet_names
# Load the first worksheet, treat -9999 as nan|, make first column (dates) the index
wb=pd.read_excel(xlsx,sheets[0],na_values=[-9999],index_col=0)
```

```python
## VIEW DATA

# Print summary of worksheet
print(wb)

# Print datatypes (dtypes) of DataFrame (wb)
print(wb.dtypes)

# Plot data, set fig size
plt.rcParams["figure.figsize"] = (20,10)
wb.plot();
```

```python
## PERFORM REGRESSION AND VIEW RESULTS FIRST TRY

# call Gates AWS the "right" value since it is our most stable and consistent measurement
# start with Gates3600 eg. x=rng['Gates AWS degC']. Make it into an array and reshape it as column
# probably a better way to do this
y=np.asarray(rng['Gates AWS degC'])
y=y.reshape(-1,1)
x1=np.asarray(rng['Gates 3600 degC'])
x1=x1.reshape(-1,1)
x2=np.asarray(rng['Kenn 6100 degC'])
x2=x2.reshape(-1,1)
x3=np.asarray(rng['Chinese AWS degC'])
x3=x3.reshape(-1,1)


# define a 1:1 line for plotting
xline=[0,20]
yline=[0,20]


# FIT AND PLOT MODELS

# Gates 3600
regr1=linear_model.LinearRegression(fit_intercept=True)
regr1.fit(x1,y)
y1_pred=regr1.predict(x1)
# save stats
print('Gates 3600 Coefficients:', regr1.coef_)
print('Gates 3600 Intercepts:', regr1.intercept_)
# save coefficients
intercepts=(regr1.intercept_)
coefs=(regr1.coef_)
print('Gates 3600 Mean squared error: %.2f'
         % mean_squared_error(x1,y1_pred))
print('Gates 3600 Coefficient of determination: %.2f'
         % r2_score(x1,y1_pred))
# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x1,y)
#
fig=plt.plot(x1,y,linewidth=0.5,linestyle=':',color='black')
fig=plt.plot(x1,y1_pred,linewidth=0.5,color='red')
# plot one to one line
plt.plot(xline,yline,linestyle='--',color='green')
ax.set_xlabel('x1')
ax.set_ylabel('y')
ax.axis('tight')

# Kenn 6100
regr2=linear_model.LinearRegression()
regr2.fit(x2,y)
y2_pred=regr2.predict(x2)
# save stats
print('Kenn 6100 Coefficients:', regr2.coef_)
print('Kenn 6100 Intercepts:', regr2.intercept_)
# save coefficients
intercepts=np.append(intercepts,regr2.intercept_)
coefs=np.append(coefs,regr2.coef_)
print('Kenn 6100 Mean squared error: %.2f'
         % mean_squared_error(x2,y2_pred))
print('Kenn 6100 Coefficient of determination: %.2f'
         % r2_score(x2,y2_pred))
# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x2,y)
fig=plt.plot(x2,y2_pred,linewidth=0.5)
# plot one to one line
plt.plot(xline,yline,linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y2')
ax.axis('tight')

# Chinese AWS
regr3=linear_model.LinearRegression()
regr3.fit(x3,y)
y3_pred=regr3.predict(x3)
# save stats
print('Chinese AWS Coefficients:', regr3.coef_)
print('Chinese AWS Intercepts:', regr3.intercept_)
# save coefficients
intercepts=np.append(intercepts,regr3.intercept_)
coefs=np.append(coefs,regr3.coef_) 
print('Chinese AWS Mean squared error: %.2f'
         % mean_squared_error(x3,y3_pred))
print('Chinese AWS Coefficient of determination: %.2f'
         % r2_score(x3,y3_pred))

# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x3,y)
fig=plt.plot(x3,y3_pred,linewidth=0.5)
# plot one to one line
plt.plot(xline,yline,linestyle='--')
ax.set_xlabel('x3')
ax.set_ylabel('y3')
ax.axis('tight')
```

```python
## PERFORM REGRESSION AND VIEW RESULTS SECOND TRY

# call Gates AWS the "right" value since it is our most stable and consistent measurement
# start with Gates3600 eg. x=rng['Gates AWS degC']. Make it into an array and reshape it as column
# probably a better way to do this
x=np.asarray(rng['Gates AWS degC'])
x=x.reshape(-1,1)
y1=np.asarray(rng['Gates 3600 degC'])
y1=y1.reshape(-1,1)
y2=np.asarray(rng['Kenn 6100 degC'])
y2=y2.reshape(-1,1)
y3=np.asarray(rng['Chinese AWS degC'])
y3=y3.reshape(-1,1)


# define a 1:1 line for plotting
xline=[0,20]
yline=[0,20]


# FIT AND PLOT MODELS

# Gates 3600
regr1=linear_model.LinearRegression()
regr1.fit(x,y1)
y1_pred=regr1.predict(x)
# save stats
print('Coefficients:', regr1.coef_)
# save coefficients
intercepts=(regr1.intercept_)
coefs=(regr1.coef_)
print('Mean squared error: %.2f'
         % mean_squared_error(x,y1_pred))
print('Coefficient of determination: %.2f'
         % r2_score(y1,y1_pred))
# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x,y1)
fig=plt.plot(x,y1_pred,linewidth=0.5)
# plot one to one line
plt.plot(xline,yline,linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y1')
ax.axis('tight')


# Kenn 6100
regr2=linear_model.LinearRegression()
regr2.fit(x,y2)
y2_pred=regr2.predict(x)
# save stats
print('Coefficients:', regr2.coef_)
# save coefficients
intercepts=np.append(intercepts,regr2.intercept_)
coefs=np.append(coefs,regr2.coef_)
print('Mean squared error: %.2f'
         % mean_squared_error(x,y2_pred))
print('Coefficient of determination: %.2f'
         % r2_score(y2,y2_pred))
# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x,y2)
fig=plt.plot(x,y2_pred,linewidth=0.5)
# plot one to one line
plt.plot(xline,yline,linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y2')
ax.axis('tight')

# Chinese AWS
regr3=linear_model.LinearRegression()
regr3.fit(x,y3)
y3_pred=regr3.predict(x)
# save stats
print('Coefficients:', regr3.coef_)
# save coefficients
intercepts=np.append(intercepts,regr3.intercept_)
coefs=np.append(coefs,regr3.coef_) 
print('Mean squared error: %.2f'
         % mean_squared_error(x,y3_pred))
print('Coefficient of determination: %.2f'
         % r2_score(y3,y3_pred))
# plot it
fig,ax=plt.subplots()
fig=plt.scatter(x,y3)
fig=plt.plot(x,y3_pred,linewidth=0.5)
# plot one to one line
plt.plot(xline,yline,linestyle='--')
ax.set_xlabel('x')
ax.set_ylabel('y3')
ax.axis('tight')
```

## Strategy

Do a regression of temps at Kenn 6100 and Gates 3600 vs: Gates AWS. multiple regression: air temp and elevation



```python
# LINEAR REGRESSION

# --- regress glacier sites with just gates AWS temps

# plot up
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(10,8))
# data points
x,y1 = homogenized.GatesAWS_degC,homogenized.Gates3600_degC
x,y2 = homogenized.GatesAWS_degC,homogenized.Kenn6100_degC
ax1.scatter(x,y1,color='darkorchid',s=20)
ax2.scatter(x,y2,color='darkorchid',s=20)

# regression 1 (Gates 3600)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y1)
r2_1 = r_value**2
xo_1 = np.linspace(x.min(),x.max(),)
Yhat_1 = (slope*xo_1) + intercept # regression y coords
ax1.plot(xo_1,Yhat_1,'-',linewidth=4,color='k')
# annotate with r2 and regression equation
ax1.annotate('$r^2$ = {:.2f}'.format(r2_1),(1,14),fontsize=12)
ax1.annotate('y = {0:.2f}(x) - {1:.2f}'.format(slope,-intercept),(1,15),fontsize=12)
# labels
ax1.set_title('Gates 3600 as a function of Gates AWS')
ax1.set_ylabel('Glacier site temp (C)',fontsize=14)
ax1.set_xlabel('Gates AWS temp (C)',fontsize=14)

# regression 2 (Kenn 6100)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y2)
r2_2 = r_value**2
xo_2 = np.linspace(x.min(),x.max(),)
Yhat_2 = (slope*xo_2) + intercept # regression y coords
ax2.plot(xo_2,Yhat_2,'-',linewidth=4,color='k')
# annotate with r2 and regression equation
ax2.annotate('$r^2$ = {:.2f}'.format(r2_2),(1,11),fontsize=12)
ax2.annotate('y = {0:.2f}(x) - {1:.2f}'.format(slope,-intercept),(1,12),fontsize=12)
# labels
ax2.set_title('Kenn 6100 as a function of Gates AWS')
ax2.set_ylabel('Glacier site temp (C)',fontsize=14)
ax2.set_xlabel('Gates AWS temp (C)',fontsize=14)

fig.tight_layout()
plt.show()
```

```python
# Gates 3600 as a function of Gates AWS
regr5=smf.ols(formula='Gates3600_degC ~ GatesAWS_degC', data=homogenized).fit()

print(regr5.params)
regr5.summary()

```

```python
# lapse rate from Tristan = -4.06 degC / km
# Elevs: Gates AWS = 1237.5, Gates3600 = 1070, ChineseAWS = 2185, Kenn6100 = 1830

lr=-4.06
# first calculate what Gates 3600 SHOULD be based on regression, then use lapse rate to calculate kenn6100 values
homogenized['Kenn6100recon'] = 1.722162+homogenized.GatesAWS_degC*0.57 + lr*(1830-1070)*.001

homogenized.head
```

```python
# Kenn 6100 as a function of ChineseAWS
regr6=smf.ols(formula='Kenn6100_degC ~ ChineseAWS_degC', data=homogenized).fit()

print(regr6.params)
regr6.summary()

```

```python
# Kenn 6100 as a function of Chinese AWS

# lapse rate from Tristan = -4.06 degC / km
# Elevs: Gates AWS = 1237.5, Gates3600 = 1070, ChineseAWS = 2185, Kenn6100 = 1830

lr=-4.06
#  calculate what Kenn6100 SHOULD be based on regression
homogenized['Kenn6100reconFROMchinese'] = homogenized.ChineseAWS_degC*0.6317+1.7617

homogenized.head
```

```python
## Compare reconstructed value of Kenn 6100 to directly measured kenn6100
# temps subplots for weekly averaged lapse rates
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,8))


# using measured kenn6100 as the x-axis
ax1.scatter(homogenized.Kenn6100_degC,homogenized.Kenn6100recon,color='steelblue',s=5)
ax1.set_title('Chinese AWS vs temps reconstructed from GatesAWS',fontsize=18)
ax1.set_ylabel('Kenn6100 reconstructed from AWS (C)',fontsize=14)
ax1.set_xlabel('Kenn6100 actual (C)',fontsize=14)

# upper site
ax2.scatter(homogenized.Kenn6100_degC,homogenized.Kenn6100reconFROMchinese,color='steelblue',s=5)
ax2.set_title('Chinese AWS vs temps reconstructed from ChineseAWS',fontsize=18)
ax2.set_ylabel('Kenn6100 reconstructed from AWS (C)',fontsize=14)
ax2.set_xlabel('Kenn6100 actual (C)',fontsize=14)



# data points
x,y3 = homogenized.Kenn6100_degC,homogenized.Kenn6100recon
x,y4 = homogenized.Kenn6100_degC,homogenized.Kenn6100reconFROMchinese

# regression 3 (Kenn6100 from GatesAWS)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y3)
print(r_value)
r2_3 = r_value**2
xo_3 = np.linspace(x.min(),x.max(),)
Yhat_3 = (slope*xo_3) + intercept # regression y coords
ax1.plot(xo_3,Yhat_3,'-',linewidth=4,color='k')
# annotate with r2 and regression equation
ax1.annotate('$r^2$ = {:.2f}'.format(r2_3),(-2.5,9),fontsize=12)
ax1.annotate('y = {0:.2f}(x) - {1:.2f}'.format(slope,-intercept),(-2.5,10),fontsize=12)
# labels
ax1.set_title('Kenn 6100 as a function of Gates AWS')
ax1.set_ylabel('Glacier site temp (C)',fontsize=14)
ax1.set_xlabel('Gates AWS temp (C)',fontsize=14)


# regression 4 (Kenn6100 from ChineseAWS)
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y4)
print(r_value)
r2_4 = r_value**2
xo_4 = np.linspace(x.min(),x.max(),)
Yhat_4 = (slope*xo_4) + intercept # regression y coords
ax2.plot(xo_4,Yhat_4,'-',linewidth=4,color='k')
# annotate with r2 and regression equation
ax2.annotate('$r^2$ = {:.2f}'.format(r2_4),(-2.5,9),fontsize=12)
ax2.annotate('y = {0:.2f}(x) - {1:.2f}'.format(slope,-intercept),(-2.5,10),fontsize=12)
# labels
ax2.set_title('Kenn 6100 as a function of Gates AWS')
ax2.set_ylabel('Glacier site temp (C)',fontsize=14)
ax2.set_xlabel('Gates AWS temp (C)',fontsize=14)

plt.show()
```

## SUMMARY

There is a good relationship between temps at Gates AWS (off-glacier) and Gates 3600 (on-glacier). There is a similar, but messier, relationship bttween temps at Chinese AWS (off-glacier) and Kenn 6100 (on-glacier). In both cases, temperatures on the glacier are significantly cooler than their off-glacier counterparts when it is warm out, but as temperatures at the off-glacier site get cooler and approach zero, the difference diminishes and then eventually reverses, to where the off-glacier site (which in both cases is at a higher elevation) becomes cooler than the on-glacier site. This relationship can't be trusted below zero C, because we don't have good data for that time period, but we don't really care because we are only interested in positive degree days, which we can calculate pretty well from the warmer days at Gates. 

Overall, Gates 3600 is on AVERAGE 2.2 degrees cooler than Gates AWS. But that said, we can model the relationship better as Gates 3600 = 0.57 x Gates AWS +1.72. This has an r2 of 0.87.

So to predict temps at Kenn 6100 (which we know for this past summer), I tried two approaches. 1) Calculate on-ice temp at Gates 3600 from the relationship above, then use the lapse rate of -4.06 degC/km (from Tristan's work) to calculate the temp there. Or 2) Calculate on-ice temp at Kenn 6100 directly from Chinese AWS. That's what I've done above, and the prediction using Gates AWS is nearly as good (r2 = 0.75, vs r2=0.77 when basing off Kenn 6100). So I think that's the way to go, except I need to do some work to re-calculate lapse rates that reflect on-ice temperatures, which I don't have a lot of good data for...



```python
# calculate best lapse rate based on gates 3600 and kenn 6100 data
# lapse rate from Tristan = -4.06 degC / km
# Elevs: Gates AWS = 1237.5, Gates3600 = 1070, ChineseAWS = 2185, Kenn6100 = 1830

homogenized['diff2'] = homogenized['Kenn6100_degC'] - homogenized['Gates3600_degC']
print ("Mean of differences between Kenn6100 and Gates3600 = ",homogenized['diff2'].mean())

# calculate lapse rate
lr2=homogenized['diff2'].mean() / (1830-1070) * 1000
print(lr2)
```
